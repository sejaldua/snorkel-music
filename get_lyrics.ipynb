```python
import requests as rq
import config
import pandas as pd 
```

```python
df_train = pd.read_csv('data/train.csv')
df_test = pd.read_csv('data/test.csv')
```

```python

# test
def get_lyrics(row):
    track_name = row['track name']
    artist = row['artist']
    lyrics = rq.get('https://api.musixmatch.com/ws/1.1/matcher.lyrics.get?' + 
                'apikey=' + config.MUSIXMATCH_API_KEY +
                '&format=json' +
                '&q_track=' + track_name +
                '&q_artist=' + artist
               )
    try:
        return dict(lyrics.json())['message']['body']['lyrics']['lyrics_body'][:-74]
    except:
        return ""
    
row = df_test.loc[5]
get_lyrics(row)
```

```python
df_train.head()
```

```python
df_train['lyrics'] = df_train.apply(get_lyrics, axis=1)
df_train.head()
```

```python
df_test['lyrics'] = df_test.apply(get_lyrics, axis=1)
df_test.head()
```

```python
df_train.to_csv('data/train-lyrics.csv', index=False)
df_test.to_csv('data/test-lyrics.csv', index=False)
```
